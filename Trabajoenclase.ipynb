# Valor Presente Neto

*El Valor Presente Neto (VPN), o Valor Actual Neto (VAN), es una **herramienta** financiera que determina la rentabilidad de un proyecto o inversión. **Calcula la diferencia** entre el valor actual de los flujos de efectivo futuros (ingresos) y la inversión inicial, descontados a una tasa de interés específica. Ayuda a decidir si un proyecto genera valor (VPN positivo) o no (VPN negativo).*

###Aspectos Claves del VPN
- **Fundamento:** Incorpora el valor del dinero en el tiempo, reconociendo que el dinero hoy vale más que la misma cantidad en el futuro.
- **Componentes:** Requiere la inversión inicial, flujos de caja proyectados y la tasa de descuento (costo de capital)
- **Interpretación:**
VPN > 0 (Positivo): El proyecto es rentable y añade valor a la empresa.

VPN < 0 (Negativo): El proyecto no es rentable, ya que los costos superan los beneficios.
